# Write a code that produces the the ** Bell numbers** count the possible partitions of a set.

**Hint**:
. Output the n-th Bell Number - [stackexchange](https://codegolf.stackexchange.com/questions/132379/output-the-n-th-bell-number)

```python
#Source: https://www.geeksforgeeks.org/bell-numbers-number-of-ways-to-partition-a-set/
def bell_numbers(n):
  # Initialize the previous row of the Bell triangle with dp[0] = 1
  dp = [1] + [0] * n

  for i in range(1, n+1):
    # The first element in each row is the same as the last element in the previous row
    prev = dp[0]
    dp[0] = dp[i-1]
    for j in range(1, i+1):
      # The Bell number for n is the sum of the Bell numbers for all previous partitions
      temp = dp[j]
      dp[j] = prev + dp[j-1]
      prev = temp

  return dp[0]

# |s| = 0 ... 69
for i in range(70):
  print(f"{i}: {bell_numbers(i)}")
```

###Write a code that that produces  list of all partitions of the set of $[n]=\{ 1,2,3,4\cdots , n-1, n\}$ with $0 \le n \le 65$ and also list all the partitions for which the sum of elements of the partition is the same constant for all.

```python
#Source: https://www.geeksforgeeks.org/generate-all-partition-of-a-set/
ap = []
def partition_set(data, index, ans):
  if index == len(data):
    # If we have considered all elements in the set, print the partition
    ap.append([ list(i) for i in ans ])
    return
    # For each subset in the partition, add the current element to it and recall
  for i in range(len(ans)):
    ans[i].append(data[index])
    partition_set(data, index + 1, ans)
    ans[i].pop()

    # Add the current element as a singleton subset and recall
  ans.append([data[index]])
  partition_set(data, index + 1, ans)
  ans.pop()


def all_partitions(data):
  ans = []
  partition_set(data, 0, ans)


if __name__ == "__main__":
  #RAM issue. For tests purrposes, n = 8.
  n = 8
  data = list(range(1, n + 1))
  all_partitions(data)
  print(f"For n = 8.\nSize of all partitions list with n: {len(ap)}. nth Bell's number: {bell_numbers(n)}")
  for l in ap:
    print(l)
```
